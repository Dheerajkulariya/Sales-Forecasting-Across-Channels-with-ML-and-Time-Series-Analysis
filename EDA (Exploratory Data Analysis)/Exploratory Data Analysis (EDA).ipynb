useful information

```python
# Copy of this dataset 
df_copy = df.copy()
df_copy
```

```python
# Change DATE column data type object type to datetime type
df_copy['DATE'] = df_copy['DATE'].astype('datetime64[ns]')
```

```python
df_copy['Year'] = df_copy['DATE'].dt.year
df_copy['Month'] = df_copy['DATE'].dt.month
df_copy['Day'] = df_copy['DATE'].dt.day
df_copy
```

```python
df_copy.dtypes
```

```python
# Total spends of all marketing channels
idx = 2
df_copy_list = ['TV_Spends','OOH_Spends','Print_Spends','FB_Spends','Search_Spends']
#df2_copy['Total_Ads_Spends'] = df2_copy[df2_copy_list].sum(axis=1)
new_col = df_copy[df_copy_list].sum(axis=1)
df_copy.insert(loc=idx, column = 'Total_Ads_Spends', value = new_col)
df_copy
```

```python
df_copy.describe().T
```

```python
print('Maximum Sale :', df_copy['Sales'].max())
print('Maximum Total_Ads_Spends :', df_copy['Total_Ads_Spends'].max())
```

```python
Sales_per_year = df_copy.groupby('Year')['Sales'].sum()   # Total sales per year
Sales_per_year
```

```python
years = [2015,2016,2017,2018,2019]
plt.figure(figsize=(12,6))
sizes = Sales_per_year
print(sizes)
labels = list(years)
print(labels)
colors = ['yellowgreen', 'lightgreen', 'darkgreen', 'gold', 'red']
plt.pie(sizes, explode=None,labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
#plt.pie(sizes, labels, colors)
plt.axis('equal')
plt.legend()
plt.show()
```

```python
Total_ads_spends_per_year = df_copy.groupby('Year')['Total_Ads_Spends'].sum()  # Total Ads spend per year
Total_ads_spends_per_year
```

```python
years = [2015,2016,2017,2018,2019]
plt.figure(figsize=(12,6))
sizes = Total_ads_spends_per_year
print(sizes)
labels = list(years)
print(labels)
colors = ['yellowgreen', 'lightgreen', 'darkgreen', 'gold', 'red']
plt.pie(sizes, explode=None,labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
#plt.pie(sizes, labels, colors)
plt.axis('equal')
plt.legend()
plt.show()
```

```python
Year = df_copy['Year']
sales = df_copy['Sales']
Total_Ads_Spends = df_copy['Total_Ads_Spends']
w = 0.4

year_bar = Year
sales_bar = [i+w for i in year_bar]
Total_Ads_Spends_bar = [i+w for i in sales_bar]

plt.figure(figsize=(12,5))
plt.xlabel('YEARS')
plt.ylabel('Sales & Total Ads Spend')
plt.title("Sales Comparision with Total Ads Spends ")
plt.ticklabel_format(style='plain') # for avoid Scientific Notation values
plt.xticks(year_bar+w/2,Year) # set position of years

plt.bar(year_bar, sales, width = w, label = 'sales')
plt.bar(sales_bar, Total_Ads_Spends, width = w, label = 'Total_Ads_Spends')
plt.legend()

#plt.ylim(0,4000000)

plt.show()
```

```python
pd.options.display.float_format = '{:.0f}'.format # for avoid Scientific Notation

# Aggregate of multiple columns
df_copy.groupby("Year")[["Sales","Total_Ads_Spends"]].agg(["mean","sum"])

#df_copy.groupby("Year").agg({"Sales":["mean","sum"],
                            #"Total_Ads_Spends":["mean","sum"]})
```

```python
# Add new column Ads_Spend_Percent Percentage of total ads spend in the sales
df_copy['Ads_Spend_Percent'] = round((df_copy['Total_Ads_Spends'] / df_copy['Sales'] * 100)).astype(str) + '%'
df_copy
#df_copy['Ads_Spend_percent'] = df_copy['Ads_Spend_percent'].astype(str) + '%'
```

```python
# Top 10 values sales wise
df_copy[['DATE','Sales','Total_Ads_Spends','Ads_Spend_Percent']].sort_values(['Sales'],ascending=False).head(10)
```

```python
# Top 10 values Total_Ads_Spends wise
df_copy[['DATE','Sales','Total_Ads_Spends','Ads_Spend_Percent']].sort_values(['Total_Ads_Spends'],ascending=False).head(10)
```

```python
df_copy_P = df_copy.copy()
df_copy_P.drop('Ads_Spend_Percent',axis = 1,inplace = True)

df_copy_P['Ads_Spend_Percent'] = round((df_copy_P['Total_Ads_Spends'] / df_copy['Sales'] * 100))  #.astype(str) + '%'

# Top 10 values Ads_Spend_Percentes wise
df_copy_P_Short = df_copy_P[['DATE','Sales','Total_Ads_Spends','Ads_Spend_Percent']].sort_values(['Ads_Spend_Percent'],ascending=False).head(10)
df_copy_P_Short['Ads_Spend_Percent'] = df_copy_P_Short['Ads_Spend_Percent'].astype(str) + '%'
df_copy_P_Short
```
